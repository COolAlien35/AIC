# AIC Artifacts

Bundles `results/`, `checkpoints/grpo/`, `exports/`, and the pipeline
logs into a single zip you can download from JupyterLab. HF Spaces
without paid persistent storage wipe `/workspace` on restart, so download
this zip before stopping the Space if you want to keep the run.

```python
import os, subprocess
REPO = '/workspace/aic-repo'
os.chdir(REPO)
subprocess.check_call([
    'zip', '-r', '/workspace/aic_artifacts.zip',
    'results', 'checkpoints/grpo', 'exports',
    'logs/grpo_progress.jsonl', 'logs/pipeline',
])
```
